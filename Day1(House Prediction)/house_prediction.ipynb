```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
data=pd.read_csv(r"C:\Users\Vimalraj\OneDrive\Desktop\30_days_Machine_learning_challenge\Day1(House Prediction)\Housing.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.replace({'yes':1,'no':0},inplace=True)
data.head()
```

```python
print(data.isnull().sum())
```

```python
data.head()
```

```python
data.replace({'furnished':2,'semi-furnished':1,'unfurnished':0},inplace=True)
```

```python
X=data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
Y=data['price']
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("MSE:",mean_squared_error(Y_test,y_pred))
print("MAE:",mean_absolute_error(Y_test,y_pred))
print("RSME:",np.sqrt(mean_squared_error(Y_test,y_pred)))
print("R-2 Score:",r2_score(Y_test,y_pred))

```

```python
plt.scatter(Y_test,y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Price")
plt.show()
```
